要素3.固有表現抽出

```python
import spacy
from spacy import displacy
import re

#モデルのインポート
nlp = spacy.load('ja_ginza')
```

```python
#ツイート文の取得
docList=[]
for i in range (len(tweetDataAll)):
    code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')
    cleaned_text = code_regex.sub('', str(tweetDataAll[i]))
    docList.append(nlp(cleaned_text))
```

```python
displacy.render(docList, style="ent", jupyter=True)
```

```python
#固有表現抽出（人名、街名ごと）
peopleList=[]
cityList=[]
cityDocList=[]
tweetNoun=[]
for doc in docList:
    for ent in doc.ents:
        tweetNoun.append(ent.text)
        if(ent.label_=="Person"):
            peopleList.append(ent.text)
        elif(ent.label_=="City"):
            cityList.append(ent.text)
            cityDocList.append(doc)
```

```python
#単語カウント
import collections

peopleListCount = collections.Counter(peopleList)
peopleListCount = sorted(peopleListCount.items(), key=lambda x:x[1] , reverse=True)

cityListCount = collections.Counter(cityList)
cityListCount = sorted(cityListCount.items(), key=lambda x:x[1] , reverse=True)

tweetNounCount = collections.Counter(tweetNoun)
tweetNounCount = sorted(tweetNounCount.items(), key=lambda x:x[1] , reverse=True)
```

```python
# 頻出単語
for person in peopleListCount[:]:
    if(person[1]<=1):
        peopleListCount.remove(person)
    elif(re.search(r'[a-zA-Z0-9]', person[0])):
        peopleListCount.remove(person)
for tn in tweetNounCount[:]:
    if(tn[1]<=5):
        tweetNounCount.remove(tn)
    elif(re.search(r'[a-zA-Z0-9]', tn[0])):
        tweetNounCount.remove(tn)
```

```python
#何が注目されているか
tweetNounList=[]
for i in tweetNounCount:
    tweetNounList.append(i[0])
tweetNounList
```

要素４, 形態素解析

```python
from janome.tokenizer import Tokenizer
import collections
import re

tokenizer = Tokenizer() 
txt = str(tweetDataAll) 
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')
cleaned_text = code_regex.sub('', txt)
noun = []
adjective = []

for token in tokenizer.tokenize(cleaned_text):
    pos = token.part_of_speech.split(',')  
    if '名詞' in pos:  
        noun.append(token.surface) 
    elif '形容詞' in pos:  
        adjective.append(token.surface) 

nounCount = collections.Counter(noun) 
nounCount = sorted(nounCount.items(), key=lambda x:x[1] , reverse=True)
for noun in nounCount[:]:
    if(re.search(r'[a-zA-Z0-9]', noun[0])):
        nounCount.remove(noun)

adjectiveCount = collections.Counter(adjective) 
adjectiveCount = sorted(adjectiveCount.items(), key=lambda x:x[1] , reverse=True)

print(adjectiveCount)
```

```python
#名詞抽出
for noun in nounCount[:]:
    if(noun[1]<=4):
        nounCount.remove(noun)
```

```python
#形容詞抽出
for adjective in adjectiveCount[:]:
    if(adjective[1]<=2):
        adjectiveCount.remove(adjective)
```

```python
#セーラームーンミュージアム
import re

tweetUserData=[]
descpNoun=[]
eventNounList=[]

for i in range(100):
    if(i==0):
        print(f'token {i+1}')
        tweets = client.search_recent_tweets(query='セーラームーンミュージアム& -is:retweet',  # 検索ワード
                                        expansions=['author_id'],
                                        user_fields=['description'],
                                        tweet_fields=['created_at','context_annotations'],
                                        max_results=100,  # 取得件数
                                        )
    else:
        print(f'token {i+1}')
        tweets = client.search_recent_tweets(query='セーラームーンミュージアム& -is:retweet',  # 検索ワード
                                        expansions=['author_id'],
                                        user_fields=['description'],
                                        tweet_fields=['created_at','context_annotations'],
                                        max_results=100,  # 取得件数
                                        next_token=token
                                        )
    
            
    for j in range (len(tweets[0])):
        id=tweets[0][j].author_id #user_id
        user = api.get_user(user_id=f"{id}")
        #ツイート文から記号の削除
        text=tweets[0][j].text.replace('\n',"")
        code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')
        cleaned_text = code_regex.sub('', text)

        eventNoun=None
        annotation=None
        adjective=None

        #descriptionの固有表現抽出
        for ent in nlpUnique(user.description).ents:
            # if(ent.label_=="Person"):
            descpNoun.append(ent.text)
        
        #ツイート文の形態素解析→形容詞抽出
        for token in tokenizer.tokenize(cleaned_text):
            pos = token.part_of_speech.split(',')  
            # if '名詞' in pos:  
            #     noun=token.surface
            #     # if(peopleListCount in noun):
            #     person=set(peopleListCount)&set(noun)

            #ここは単語ベクトルで置き換えたい（形容詞は事前定義）
            if '形容詞' in pos:
                if('良' in token.surface):
                    adjective="良い"
                elif('可愛' in token.surface) or ('かわ' in token.surface):
                    adjective="可愛い"
                elif('美味' in token.surface):
                    adjective="美味しい"
                elif('すご' in token.surface):
                    adjective="すごい"
                else:
                    adjective=token.surface

        #ツイート文の固有表現抽出
        if('代行' in cleaned_text):
            continue
        elif('郵送' in cleaned_text):
            continue
        elif('譲' in cleaned_text):
            continue
        elif('送料' in cleaned_text):
            continue
        elif('手渡し' in cleaned_text):
            continue
        if(cleaned_text):
            for ent in nlpUnique(cleaned_text).ents:
                # if(ent.label_=="Person"):
                if(bool(re.search(r'[a-zA-Z0-9]',ent.text))==False):
                    eventNoun=ent.text
                    eventNounList.append(eventNoun)
        
        #ツイート文のアノテーション
        if(tweets[0][j].context_annotations):
            annotation=tweets[0][j].context_annotations[0]['entity']['id']

        #ユーザデータの統合
        tweetUserData.append([
            id,
            user.description,
            user.location,
            user.followers_count,
            user.friends_count,
            user.created_at,
            user.favourites_count,
            user.verified,
            annotation, #ツイートアノテーション
            text,
            nlp(text)[0]['label'],
            eventNoun,
            # modelWords[noun],
            adjective,
            tweets[0][j].created_at.strftime('%Y-%m-%d %H')
            ])
    if('next_token' not in tweets.meta):
        break
    token=tweets.meta['next_token']
```

```python
#イベント来場者のクラスタ分析
from sklearn.cluster import KMeans
from sklearn import datasets, preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from matplotlib import cm

# pd_df = dfUserInfo.drop(['user_id','text'], axis=1)
pd_df = dfUserInfo.drop(['user_id','text','created_at','tweet_time'], axis=1)
pd_df = pd.get_dummies(pd_df)
sc = StandardScaler()
df_sc = sc.fit_transform(pd_df)
df_sc = pd.DataFrame(pd_df, columns=pd_df.columns)
model = KMeans(n_clusters=3, random_state=1)
model.fit(df_sc)
```
